```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(20)
y=np.random.normal(10,1,20)
z=np.random.normal(10,2,20)
k=np.random.normal(10,3,20)
```

```python
plt.plot(x,y,color='r')
plt.plot(x,z,color='g')
plt.plot(x,k);# sonda yazilan ;  veri detayinin cikmasini engeller
```

```python
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot(131)
ax.plot(x,y)

bx=fig.add_subplot(132)
bx.plot(x,z,color='r')

cx=fig.add_subplot(133)
cx.plot(x,k);
```

```python
fig=plt.figure(figsize=(10,15))
ax=fig.add_subplot(311)
ax.set_title("x in y e gore degisimi")
ax.set_xlabel("x verileri")
ax.set_ylabel("y verileri")
ax.plot(x,y)

bx=fig.add_subplot(312)
bx.set_title("x in z e gore degisimi")
bx.set_xlabel("z verileri")
bx.plot(x,z)

cx=fig.add_subplot(313)
cx.set_title("x in k ya gore degisimi")
cx.set_xlabel("k verileri")
cx.plot(x,k,color='g');
```

```python
#Feature Engineering (oznitelik muhendisligi)- verinin aslinda olmayan ancak veriden yararlanilip uretilen yeni sutun islemine denir
```

```python
df['Date']=pd.to_datetime(df['Date']) #tarihi yazi formatindan tarih formatina cevirdik
```

```python
df.info()
```

```python
df['day']=(df['Date']).dt.day
df['month']=(df['Date']).dt.month
df['year']=(df['Date']).dt.year #tarihi gun ay yil olarak sutunlara ayirdik
```

```python
df['month_name']=df['Date'].dt.month_name()
df['weekday']=df['Date'].dt.day_name() #ay gun isimlerine cevirdik
```

```python
df['Time']=pd.to_datetime(df['Time'])
```

```python
df['Hour']=(df['Time']).dt.hour
```

### Feature Engineering

```python
#Burada data uzerinde manipulasyon yapip ekleme cikarma yapabiliriz
```

```python
df['Attrition'] #Bunu rakama cevirmemiz lazim
```

```python
# dictionary: sozluk
#kelime aciklama-tanimlama kullanrrak kelimeyi pythonda degistireegiz
d={"Yes":1,"No":0}
```

```python
df['Attrition']=df['Attrition'].map(d)
# map for dongusunun modern hali
```

```python
df['Attrition']
```

```python
#df['Attrition']=df['Attrition'].astype('int64') #ondalik kismi tam sayi olarak yazmak istedik
```

```python
# Simdi Attribute teki bos verileri doldurmamiz gerek bunun icin dolulara bakarak boslari tahmin edecegiz.
```

```python
bos=df[df['Attrition'].isnull()]
```

```python
dolu=df[df['Attrition'].notnull()] #bos ve dolu olan sutunlara gore ayirdik
```

```python
x=dolu.drop('Attrition',axis=1) #doludan boslari cikardik
y=dolu['Attrition']
```

```python
x.sample(5)
```

```python
x.shape
```

```python
x['BusinessTravel'].value_counts()
```

```python
x=pd.get_dummies(x,drop_first=True) # yazilari sayilara cevirdik
```

```python
x.head()
```

```python
x.shape
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
g=GaussianNB()
b=BernoulliNB()
# 2 tane classification modeli cagirdik
```

```python
g.fit(x,y)
b.fit(x,y)
```

```python
tahmin1=g.predict(x)
tahmin2=b.predict(x)
```

```python
# classification Accuracy, Precision, Recali, f1 score ve confuson matrix
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
accuracy_score(tahmin1,y)
```

```python
confusion_matrix(tahmin1,y)
```

```python
print(classification_report(tahmin1,y))
```

```python
sns.heatmap(confusion_matrix(tahmin1,y),annot=True);
```

```python
accuracy_score(tahmin2,y)
```

```python
confusion_matrix(tahmin2,y)
```

```python
print(classification_report(tahmin2,y))
```

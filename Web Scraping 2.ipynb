```python
url="https://www.karar.com/yazarlar/zafer-acar"
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
soup.prettify
```

```python
linkler=soup.find_all("article",{"class":"item box-shadow"})
```

```python
for i in linkler:
    print(i)
```

```python
yazilar=[]
basliklar=[]
for link in linkler:
    yazilinki="https://www.karar.com"+link.a.get("href")
    #print(yazilinki)
    page=requests.get(yazilinki)
    soup=BeautifulSoup(page.content,'html.parser')
    baslik=soup.find('h1',{"class":"content-title"}).text
    yazi=soup.find('article',{"class":"article-content"}).text
    yazilar.append(yazi)
    basliklar.append(baslik)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=["konu","makale"])
```

```python
df["konu"]=basliklar
df["makale"]=yazilar
```
